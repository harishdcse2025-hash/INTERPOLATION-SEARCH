```python
import random

def binary_search(arr, target):
    """
    Implements Binary Search and counts the number of probes.
    """
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high:
        probes += 1
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid, probes
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1, probes


def interpolation_search(arr, target):
    """
    Implements Interpolation Search and counts the number of probes.
    """
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and arr[low] <= target <= arr[high]:
        probes += 1

        # Prevent division by zero
        if arr[high] == arr[low]:
            if arr[low] == target:
                return low, probes
            break

        # Estimate the position
        pos = low + int(
            ((high - low) * (target - arr[low])) /
            (arr[high] - arr[low])
        )

        if arr[pos] == target:
            return pos, probes
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1

    return -1, probes


# ---------------- Main Program ----------------
if __name__ == "__main__":

    # Generate sorted unique student roll numbers
    array_size = 5000
    student_rolls = sorted(random.sample(range(1, 10001), array_size))

    # Select a roll number that definitely exists
    target_index = random.randint(0, array_size - 1)
    target_roll = student_rolls[target_index]

    print(f"Array Size (Total Students): {array_size}")
    print(f"Searching for Roll Number: {target_roll}")
    print("-" * 40)

    # Binary Search
    bs_idx, bs_probes = binary_search(student_rolls, target_roll)

    print("Binary Search")
    print(f"Found at index : {bs_idx}")
    print(f"Probes required: {bs_probes}")

    print("-" * 40)

    # Interpolation Search
    is_idx, is_probes = interpolation_search(student_rolls, target_roll)

    print("Interpolation Search")
    print(f"Found at index : {is_idx}")
    print(f"Probes required: {is_probes}")

    print("-" * 40)

    # Comparison
    if bs_probes < is_probes:
        print("Binary Search used fewer probes.")
    elif is_probes < bs_probes:
        print("Interpolation Search used fewer probes.")
    else:
        print("Both algorithms used the same number of probes.")
```
